`````python
# Das ist ein Kommentar in einer CodeZelle
ich_bin_eine_variable = "ich bin eine Variable" # ein String in einer Variablen
ich_bin_auch_eine_variable = 15                 # ein Integer in einer Variablen
ich_bin_eine_liste = [1,2,3,4,5,6]              # eine Liste mit Integers
ich_bin_ein_dictionary = {"eins": 1, "zwei": 2} # ein Dictionary
`````

`````python
# Ohne print
ungerade = []
for i in range(1,20):
    zahl = 2*i-1
    ungerade.append(zahl)
    
ungerade
`````

`````python
# Eine Funktion
def is_prime(n):
    for i in range(2,int(n/2)):
        if (n%i) == 0:
            return False
        return True
`````

`````python
for zahl in ungerade:
    print(zahl, is_prime(zahl))
`````

`````python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn-notebook")

x = np.random.randint(0, 100, size=100)
y = np.random.randint(1, 100, size=100)

_,ax = plt.subplots()
ax.plot(x,y)
ax.set(ylabel="Zufall", xlabel="Auch Zufall");
`````

`````python
_,ax = plt.subplots()
ax.scatter(x,y)
ax.set(ylabel="Zufall", xlabel="Auch Zufall")
`````
